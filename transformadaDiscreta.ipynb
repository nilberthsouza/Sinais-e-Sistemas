```python
import numpy as np
import matplotlib.pyplot as plt

def dft(x):
    N = len(x)
    n = np.arange(N)
    k = n.reshape((N, 1))
    e = np.exp(-2j * np.pi * k * n / N)
    return np.dot(e, x)

N = 100  # Número de amostras
n = np.arange(N)  # Vetor de amostras

x = np.sin(np.pi/8 * n) - 2*np.cos(np.pi/4 * n)
X = dft(x)

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.stem(n, x)
plt.title('Sinal x[n]')
plt.xlabel('n')
plt.ylabel('Amplitude')

freq = np.arange(N) / N
plt.subplot(1, 2, 2)
plt.stem(freq, np.abs(X))
plt.title('Transformada Discreta de Fourier (TDF)')
plt.xlabel('Frequência normalizada')
plt.ylabel('Magnitude')
plt.tight_layout()
plt.show()
```
