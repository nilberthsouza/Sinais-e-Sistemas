```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def pulso_retangular(t,T):
  return np.where(np.abs(t) <= T/2,1,0)
```

```python
t = np.linspace(-10,10,500)
T = 5

sinal1 = pulso_retangular(t,T)
sinal2 = pulso_retangular(t,T)
resultado = np.convolve(sinal1,sinal2,'same')
```

```python
plt.plot(t,sinal1,'g',linewidth=2)
plt.title("Sinal1")
plt.xlabel("Tempo")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
plt.plot(t,sinal1,'b',linewidth=2)
plt.title("Sinal2")
plt.xlabel("Tempo")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
plt.plot(t,resultado,'r',linewidth=2)
plt.title("Convolução de dois pulsos retangulares")
plt.xlabel("Tempo")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
frequencia = 1  # Frequência em Hz
amplitude = 0.5  # Amplitude do sinal senoidal

sinal3 = amplitude * np.sin(2 * np.pi * frequencia * t)

plt.plot(t,sinal3,'g',linewidth=2)
plt.title("Sinal1")
plt.xlabel("Tempo")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python

transformada = np.fft.fft(sinal3)

frequencias = np.fft.fftfreq(len(t), t[1] - t[0])

plt.figure(figsize=(10, 6))
plt.plot(frequencias, np.abs(transformada))
plt.xlabel('Frequência (Hz)')
plt.ylabel('Amplitude da Transformada')
plt.title('Espectro de Frequência do Sinal Senoidal')
plt.grid(True)
plt.show()
```
