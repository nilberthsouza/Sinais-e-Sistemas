```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros da onda senoidal
frequency = 2500  # Frequência em Hz (2.5 kHz)
amplitude_pp = 4  # Amplitude pico a pico em V

# Tempo de amostragem
sampling_rate = 100000  # Taxa de amostragem em Hz
sampling_time = 1 / sampling_rate  # Período de amostragem

# Tempo total da gravação
duration = 0.01  # 10 ms

# Geração do sinal senoidal
t = np.arange(0, duration, sampling_time)
x = amplitude_pp * np.sin(2 * np.pi * frequency * t)

# Plot do sinal senoidal gerado
plt.figure()
plt.plot(t, x)
plt.title('Onda Senoidal Gerada')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude (V)')
plt.grid(True)
plt.show()
```

```python

# Cálculo da FFT
fft_freq = np.fft.fftfreq(len(x), d=sampling_time)
fft_signal = np.fft.fft(x)

# Plot da magnitude da FFT
plt.figure()
plt.plot(fft_freq, np.abs(fft_signal))
plt.title('Magnitude da FFT')
plt.xlabel('Frequência (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

O sinal gerado é uma onda senoidal de 2.5 kHz com uma amplitude de 4 Vpp. Em seguida, plotamos o sinal gerado no domínio do tempo e a magnitude da FFT no domínio da frequência.

A análise dos resultados obtidos envolve examinar o gráfico da magnitude da FFT. Você deve observar um pico na frequência de 2.5 kHz, que é a frequência da onda senoidal que geramos. Além disso, é possível que haja componentes em outras frequências, devido a imperfeições na geração do sinal e à resolução finita da FFT. No entanto, a frequência dominante deve ser a frequência da onda senoidal original.

Lembre-se de que o tamanho da FFT, a taxa de amostragem e a duração do sinal podem afetar a resolução da análise em frequência. Quanto maior o tamanho da FFT e a taxa de amostragem, maior será a resolução.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros da onda quadrada
frequency = 2500  # Frequência em Hz (2.5 kHz)
amplitude_pp = 4  # Amplitude pico a pico em V

# Tempo de amostragem
sampling_rate = 100000  # Taxa de amostragem em Hz
sampling_time = 1 / sampling_rate  # Período de amostragem

# Tempo total da gravação
duration = 0.01  # 10 ms

# Geração do sinal quadrado
t = np.arange(0, duration, sampling_time)
x = amplitude_pp * np.sign(np.sin(2 * np.pi * frequency * t))

# Plot do sinal quadrado gerado
plt.figure()
plt.plot(t, x)
plt.title('Onda Quadrada Gerada')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude (V)')
plt.grid(True)
plt.show()

# Cálculo da FFT
fft_freq = np.fft.fftfreq(len(x), d=sampling_time)
fft_signal = np.fft.fft(x)

# Plot da magnitude da FFT
plt.figure()
plt.plot(fft_freq, np.abs(fft_signal))
plt.title('Magnitude da FFT')
plt.xlabel('Frequência (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

Ao analisar a magnitude da FFT, você verá componentes em várias frequências. A frequência dominante será a frequência fundamental (2.5 kHz) da onda quadrada. Além disso, você observará a presença de harmônicos ímpares (5 kHz, 7.5 kHz, etc.), característicos da forma de onda quadrada.

Lembre-se de que, devido à natureza do espectro de frequência de uma onda quadrada, haverá uma série de componentes espalhados ao longo das frequências. Isso ocorre devido à rápida transição da onda quadrada entre os níveis altos e baixos, que gera uma ampla gama de frequências.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros da onda triangular
frequency = 2500  # Frequência em Hz (2.5 kHz)
amplitude_pp = 4  # Amplitude pico a pico em V

# Tempo de amostragem
sampling_rate = 100000  # Taxa de amostragem em Hz
sampling_time = 1 / sampling_rate  # Período de amostragem

# Tempo total da gravação
duration = 0.01  # 10 ms

# Geração do sinal triangular
t = np.arange(0, duration, sampling_time)
x = amplitude_pp * (2 * np.abs((t * frequency) % 1) - 1)

# Plot do sinal triangular gerado
plt.figure()
plt.plot(t, x)
plt.title('Onda Triangular Gerada')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude (V)')
plt.grid(True)
plt.show()

# Cálculo da FFT
fft_freq = np.fft.fftfreq(len(x), d=sampling_time)
fft_signal = np.fft.fft(x)

# Plot da magnitude da FFT
plt.figure()
plt.plot(fft_freq, np.abs(fft_signal))
plt.title('Magnitude da FFT')
plt.xlabel('Frequência (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

Neste código, geramos uma onda triangular utilizando uma função que gera uma onda triangular a partir da parte positiva de um sinal de dente de serra. A frequência da onda é de 2.5 kHz e a amplitude é de 4 Vpp. Em seguida, calculamos a FFT e plotamos a magnitude da FFT no domínio da frequência.

Ao analisar a magnitude da FFT, você observará componentes em várias frequências, incluindo a frequência fundamental (2.5 kHz) da onda triangular. Além disso, você verá a presença de harmônicos ímpares (7.5 kHz, 12.5 kHz, etc.), característicos da forma de onda triangular.

Lembre-se de que, assim como na onda quadrada, a natureza da forma de onda triangular resultará em componentes espalhados ao longo das frequências no espectro de frequência. Isso ocorre devido às transições suaves entre os níveis altos e baixos da onda triangular.
