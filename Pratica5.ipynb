```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros da onda senoidal
frequency = 2500  # Frequncia em Hz (2.5 kHz)
amplitude_pp = 4  # Amplitude pico a pico em V

# Tempo de amostragem
sampling_rate = 100000  # Taxa de amostragem em Hz
sampling_time = 1 / sampling_rate  # Perodo de amostragem

# Tempo total da gravao
duration = 0.01  # 10 ms

# Gerao do sinal senoidal
t = np.arange(0, duration, sampling_time)
x = amplitude_pp * np.sin(2 * np.pi * frequency * t)

# Plot do sinal senoidal gerado
plt.figure()
plt.plot(t, x)
plt.title('Onda Senoidal Gerada')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude (V)')
plt.grid(True)
plt.show()
```

```python

# Clculo da FFT
fft_freq = np.fft.fftfreq(len(x), d=sampling_time)
fft_signal = np.fft.fft(x)

# Plot da magnitude da FFT
plt.figure()
plt.plot(fft_freq, np.abs(fft_signal))
plt.title('Magnitude da FFT')
plt.xlabel('Frequncia (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

O sinal gerado  uma onda senoidal de 2.5 kHz com uma amplitude de 4 Vpp. Em seguida, plotamos o sinal gerado no domnio do tempo e a magnitude da FFT no domnio da frequncia.

A anlise dos resultados obtidos envolve examinar o grfico da magnitude da FFT. Voc deve observar um pico na frequncia de 2.5 kHz, que  a frequncia da onda senoidal que geramos. Alm disso,  possvel que haja componentes em outras frequncias, devido a imperfeies na gerao do sinal e  resoluo finita da FFT. No entanto, a frequncia dominante deve ser a frequncia da onda senoidal original.

Lembre-se de que o tamanho da FFT, a taxa de amostragem e a durao do sinal podem afetar a resoluo da anlise em frequncia. Quanto maior o tamanho da FFT e a taxa de amostragem, maior ser a resoluo.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros da onda quadrada
frequency = 2500  # Frequncia em Hz (2.5 kHz)
amplitude_pp = 4  # Amplitude pico a pico em V

# Tempo de amostragem
sampling_rate = 100000  # Taxa de amostragem em Hz
sampling_time = 1 / sampling_rate  # Perodo de amostragem

# Tempo total da gravao
duration = 0.01  # 10 ms

# Gerao do sinal quadrado
t = np.arange(0, duration, sampling_time)
x = amplitude_pp * np.sign(np.sin(2 * np.pi * frequency * t))

# Plot do sinal quadrado gerado
plt.figure()
plt.plot(t, x)
plt.title('Onda Quadrada Gerada')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude (V)')
plt.grid(True)
plt.show()

# Clculo da FFT
fft_freq = np.fft.fftfreq(len(x), d=sampling_time)
fft_signal = np.fft.fft(x)

# Plot da magnitude da FFT
plt.figure()
plt.plot(fft_freq, np.abs(fft_signal))
plt.title('Magnitude da FFT')
plt.xlabel('Frequncia (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

Ao analisar a magnitude da FFT, voc ver componentes em vrias frequncias. A frequncia dominante ser a frequncia fundamental (2.5 kHz) da onda quadrada. Alm disso, voc observar a presena de harmnicos mpares (5 kHz, 7.5 kHz, etc.), caractersticos da forma de onda quadrada.

Lembre-se de que, devido  natureza do espectro de frequncia de uma onda quadrada, haver uma srie de componentes espalhados ao longo das frequncias. Isso ocorre devido  rpida transio da onda quadrada entre os nveis altos e baixos, que gera uma ampla gama de frequncias.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros da onda triangular
frequency = 2500  # Frequncia em Hz (2.5 kHz)
amplitude_pp = 4  # Amplitude pico a pico em V

# Tempo de amostragem
sampling_rate = 100000  # Taxa de amostragem em Hz
sampling_time = 1 / sampling_rate  # Perodo de amostragem

# Tempo total da gravao
duration = 0.01  # 10 ms

# Gerao do sinal triangular
t = np.arange(0, duration, sampling_time)
x = amplitude_pp * (2 * np.abs((t * frequency) % 1) - 1)

# Plot do sinal triangular gerado
plt.figure()
plt.plot(t, x)
plt.title('Onda Triangular Gerada')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude (V)')
plt.grid(True)
plt.show()

# Clculo da FFT
fft_freq = np.fft.fftfreq(len(x), d=sampling_time)
fft_signal = np.fft.fft(x)

# Plot da magnitude da FFT
plt.figure()
plt.plot(fft_freq, np.abs(fft_signal))
plt.title('Magnitude da FFT')
plt.xlabel('Frequncia (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

Neste cdigo, geramos uma onda triangular utilizando uma funo que gera uma onda triangular a partir da parte positiva de um sinal de dente de serra. A frequncia da onda  de 2.5 kHz e a amplitude  de 4 Vpp. Em seguida, calculamos a FFT e plotamos a magnitude da FFT no domnio da frequncia.

Ao analisar a magnitude da FFT, voc observar componentes em vrias frequncias, incluindo a frequncia fundamental (2.5 kHz) da onda triangular. Alm disso, voc ver a presena de harmnicos mpares (7.5 kHz, 12.5 kHz, etc.), caractersticos da forma de onda triangular.

Lembre-se de que, assim como na onda quadrada, a natureza da forma de onda triangular resultar em componentes espalhados ao longo das frequncias no espectro de frequncia. Isso ocorre devido s transies suaves entre os nveis altos e baixos da onda triangular.
