```python
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt

data = sio.loadmat('arquivo.mat')
sinal = data['sinal'].flatten()
```

```python

duracao = len(sinal) / 2000

tam_janela = 1000

energia = np.zeros(len(sinal) - tam_janela)
```

```python

for c in range(len(sinal) - tam_janela):
    energia[c] = np.sum(sinal[c:c + tam_janela] ** 2)
```

```python
plt.figure()
plt.plot(sinal, 'k-')
plt.plot(range(tam_janela, len(sinal)), energia, 'r-')
plt.xlabel('Amostra')
plt.ylabel('Energia')
plt.legend(['Sinal de Audio', 'Energia'])
plt.title('Sinal de Audio e Energia')
plt.grid(True)
plt.show()
```

```python
plt.figure()
plt.plot(range(tam_janela, len(sinal)), energia, 'r-')
plt.xlabel('Amostra')
plt.ylabel('Energia')
plt.title('Sinal de Energia')
plt.grid(True)
plt.show()
```

```python
limiar = 15

indices = np.where(energia > limiar)[0]

n_batimentos = len(np.where(np.diff(indices) > 1)[0]) + 1

n_bpm = 60 * n_batimentos / duracao

print("numero de respiraes:", n_batimentos)
print("incurses por minuto):", n_bpm)
```
